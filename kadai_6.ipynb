```python
from datetime import datetime

class Event:
    # コンストラクタ
    def __init__(self, name, date, location):
        self.name = name
        self.date = date  # datetime型で受け取る
        self.location = location

    # 開催日までの日数を返すメソッド
    def days_until(self):
        today = datetime.now()
        delta = self.date - today
        return delta.days


# ===== インスタンス作成 =====
# 未来の日付を設定（例：2026年12月1日）
event_date = datetime(2026, 12, 1)

event = Event(
    "プログラミング初心者向けセミナー",
    event_date,
    "東京国際フォーラム"
)

# ===== 各属性の出力 =====
print("イベント名:", event.name)
print("開催日:", event.date.strftime("%Y-%m-%d"))
print("開催場所:", event.location)
```

```python
class Human:
    # コンストラクタ（初期化）
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # nameとageを出力するメソッド
    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)


# インスタンスを変数に代入
person = Human("山田太郎", 25)

# メソッドを使って出力
person.printinfo()
```
